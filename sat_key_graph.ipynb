```
#reading satoshi pubx
filepath = "/content/drive/MyDrive/btc_datas/all.csv"
fin = open(filepath, "r")
data = fin.read()
fin.close()
data=data.split("\n")

_arx=[]
_ary=[]



for i in data[:]:
  if i!='':
    pub=i.split(",")[0]
    pubx=pub[2:66]
    puby=pub[66:]
    
    # print(pub)
    # print pubx
    # print puby
    ipubx=int("0x"+pubx,16)
    ipuby=int("0x"+puby,16)
    _arx.append(ipubx)
    _ary.append(ipuby)
```
